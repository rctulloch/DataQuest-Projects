```python
# drop all three of these as they don't add anything
autos = autos.drop(['num_pictures','seller','offer_type'], axis=1)
```

### clean column data types

```python
# convert price to an integer
autos['price'] = (autos['price']
                      .str.replace('$','')
                      .str.replace(',','')
                      .astype(int)
                 )
```

```python
# convert odometer to an integer
autos['odometer'] = (autos['odometer']
                         .str.replace('km','')
                         .str.replace(',','')
                         .astype(int)
                    )
```

```python
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
```

**registration_year** after 2016 seems unlikely as the data was crawled in 2016, selling cars from the future seems improbable. and given that the first practicle cars weren't produced until the 1800s, any registration year prior to this also seems implausible.

### Cleaning Registration Year
lets look at the impact of removing values before 1900 and after 2016 as discussed above

```python
(~autos["registration_year"].between(1900,2016)).value_counts(normalize=True)*100
```

as less than 4% of the data falls into these registration years, we will remove them.

```python
autos = autos[(autos["registration_year"].between(1900,2016))]
autos['registration_year'].describe()
```

### Investigation the Brand Column
Implimenting Aggregation

```python
autos.brand.unique()
```

```python
autos.brand.value_counts(normalize=True) *100
```

```python
(autos.brand.value_counts(normalize=True).head() *100).sum()
```

by taking just the top 5 most common brands we will cover over 60% of the data so we will aggregate on these.

```python
mean_prices_per_brand = {}
top5_brands = autos.brand.value_counts().head().index
for brand in top5_brands:
    mean_prices_per_brand[brand] = int(autos.loc[autos['brand'] == brand,'price'].mean())

mean_prices_per_brand
```

we can clearly see that opel is the cheapest brand and that audi, bmw and mercedes_benz are more expensive brands
volkswagen appears to be a more mid-range option

### Average Mileage across top 5 brands

```python
brands = pd.Series(mean_prices_per_brand)
brands = pd.DataFrame(brands, columns=['mean_price'])
brands
```

```python
mean_mileage_per_brand = {}
for brand in top5_brands:
    mean_mileage_per_brand[brand] = int(autos.loc[autos['brand'] == brand,'odometer_km'].mean())
    
mean_mileage = pd.Series(mean_mileage_per_brand)
brands['mean_mileage'] = mean_mileage
brands
```

the mean mileages do not vary significantly across brands whereas the mean price does.
